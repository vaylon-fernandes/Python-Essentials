**Q3**: Write a program for printing the tables from 1,10 using Nested For Loop

````python
for number in range(1, 11): 
    for multiplier in range(1, 11):
        product = number * multiplier
        print(f"{number} x {multiplier} = {product}")
    print()
````

**Q4**: Write a program to Print X Prime Numbers using While Loop starting from 0, and take the INput of X from
the user 

````python
x = int(input("How many prime numbers do you want ?\n"))

primes = [2]

check_num = 3
while len(primes) < x:
    
    for factor in range(3,check_num,2):
        if check_num % factor == 0:
            check_num += 2
            break
    else:
        primes.append(check_num)
        check_num+=2
print(primes)
````

**Explaination**:<br> 
Since, 2 is the smallest prime number, we store it in the list of ```primes```. To get **X** primes, we compare the length of the ```primes``` list with the user input ```x```  as a condition for the while loop. Since, all even numbers are multiples of 2, they  cannot be prime numbers. Hence, we can check only odd numbers in the range of 3 to the current number given by ```check_num```. <br>
The code block <br> 
```python 
for factor in range(3,check_num,2):
        if check_num % factor == 0:
            check_num += 2
            break ```
Checks if the current number is prime. If the number happens to be composite, the ```if``` statement condition gets satisfied, ```check_num``` is increemented by 2 (for the next odd number) and also the  ```for``` loop is terminated and the  ```while``` loop continues and then the ```for``` loop checks the next number.<br> If the number happens to the be a prime, the ```for``` loop exits and the ```else``` block appends the current number to the ```primes``` list.
